```python
#--------------------------------------------
#카메라 영상처리
```

```python
import cv2
cap=cv2.VideoCapture(0) #0번 default 카메라 설정
cap=cv2.VideoCapture('gyeongju.mov')
cap.set(1,1)
#cap.resize('gyeongju.mov',dsize=(0,0),fx=0.5,fy=0.5,interpolation=cv2.INTER_LINER)
#동영상을 가져오고싶으면 0에다가 동영상파일을 넣으면 된다
```

```python
while cap.isOpened():
    #카메라 프레임 읽기
    success,frame=cap.read()
    if success :
        #프레임 출력
        cv2.imshow('Camera Window',frame)
        
        #ESC를 누르면 종료
        key = cv2.waitKey(1) & 0xFF
        if (key==27):
            break

cap.release()
cv2.destroyAllWindows()
```
